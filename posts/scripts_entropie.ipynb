```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generer_distribution_pile_ou_face(nombre_tirages,nombre_pieces):
    """
    Fonction qui réalise un nombre de tirages donné
    avec un nombre de pièces donné
    Input
        nombre_tirages : nombre de tirages
        nombre_pieces : nombre de pièces par tirage
    Return
        x : nombre de tirages pour un nombre de pièces sur Pile donné
        bins : nombre d'arrangements non ordonnés
        xticks : valeur à écrire sur l'axe des abcisses
    
    """
    x = np.random.choice([0,1],[nombre_tirages,nombre_pieces]).sum(axis=1)
    bins = np.arange(-0.5,nombre_pieces+1.5,1)
    xticks = np.arange(0,nombre_pieces+1)
    return x,bins,xticks
```

## Tirages avec 1 pièce

```python
nbr_tirages = 10000 # 10000 tirages
nbr_pieces = 1 # 1 pièce
x,bins,xticks = generer_distribution_pile_ou_face(nbr_tirages,nbr_pieces)
plt.xticks(xticks,['Face','Pile'],fontsize=14)
plt.yticks(fontsize=12)
plt.hist(x,bins=bins,align='mid',edgecolor='black',alpha=0.75,rwidth=0.8)
plt.ylabel('Nombre de tirages',fontsize=14)
plt.show
#plt.savefig('LancersParTirage_1.png',bbox_inches='tight')
unique, counts = np.unique(x, return_counts=True)
print('Nombre de tirages pour chaque arrangement non ordonné :')
dict(zip(unique, counts))
```
